```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler # used for encoding categorical data
from sklearn.model_selection import train_test_split # used for splitting training and testing data
from sklearn.preprocessing import StandardScaler, MinMaxScaler # used for feature scaling
import matplotlib.pyplot as plt#used for displaying the heat map
import seaborn as sns#used for displaying the heat map
from sklearn.utils import resample #used for resampling
from sklearn.ensemble import RandomForestClassifier #used for random forest
from sklearn.neighbors import KNeighborsRegressor
import csv
```

### Reading The .CSV

```python
#Reading dataset
dataset = pd.read_csv('train.csv') 
#Displat dataset shape
dataset.shape
#Display first few rows
dataset.head(30)
```

```python
dataset.info()
```

### Dropping not needed columns

```python
dataset.drop('Reason', axis='columns', inplace=True)
dataset.drop('Additinal Info', axis='columns', inplace=True)
numiricColumns = dataset.select_dtypes(include=['int64', 'float64'])
for column in numiricColumns.columns:
    dataset.drop(column, axis='columns', inplace=True)
dataset
```

### Encoding The Table

```python
labelencoder = LabelEncoder()

for feture in dataset.columns.values:
    dataset[feture] = labelencoder.fit_transform(dataset[feture])
    

dataset = pd.concat([dataset, numiricColumns], axis=1)
dataset.head(30)
```

### Splitting Into Clusters

```python
reject_dataset = dataset[dataset.state == 1]
accept_dataset = dataset[dataset.state == 0]
```

```python
reject_unsampled = resample(reject_dataset, 
                                 replace=True,     # sample with replacement
                                 n_samples=288690,    # to match majority class
                                 random_state=123) # reproducible results
reject_unsampled
```

```python
upsampledDataset = pd.concat([reject_unsampled, accept_dataset])
upsampledDataset
```

```python
y_train = upsampledDataset['state']
upsampledDataset.drop('state', axis='columns', inplace=True)
x_train = upsampledDataset
x_train
```

### Testing

```python
#Reading dataset
test_dataset = pd.read_csv('test.csv') 

#Displat dataset shape
test_dataset.shape
test_dataset.drop('Reason', axis='columns', inplace=True)
test_dataset.drop('Additinal Info', axis='columns', inplace=True)

test_numiricColumns = test_dataset.select_dtypes(include=['int64', 'float64'])
for column in test_numiricColumns.columns:
    test_dataset.drop(column, axis='columns', inplace=True)

labelencoder = LabelEncoder()

for feture in test_dataset.columns.values:
    test_dataset[feture] = labelencoder.fit_transform(test_dataset[feture])


test_dataset = pd.concat([test_dataset, test_numiricColumns], axis=1)


RF_clf = RandomForestClassifier(n_estimators = 10, random_state=0).fit(x_train, y_train)

a = RF_clf.predict(test_dataset)

id_list =test_dataset['Unnamed: 0'].values

with open('ans.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['ID', 'state'])
    for i in range(len(a)):
        if a[i] == 0:
            writer.writerow([id_list[i],"Accepted"])
        else:
            writer.writerow([id_list[i],"rejected"])
```
